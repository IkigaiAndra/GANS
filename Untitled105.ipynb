```python
# Training Loop
for epoch in range(epochs):
    for i, (real_images, _) in enumerate(train_loader):
        real_images = real_images.to(device)

        # Train Critic (Discriminator)
        optimizer_c.zero_grad()

        # Generate fake images
        z = torch.randn(real_images.size(0), z_dim).to(device)
        fake_images = generator(z)

        # Get critic scores for real and fake images
        critic_real = critic(real_images)
        critic_fake = critic(fake_images.detach())

        # Compute Wasserstein loss for critic
        loss_critic = -(critic_real.mean() - critic_fake.mean())

        # Compute gradient penalty
        gp = compute_gradient_penalty(critic, real_images, fake_images)
        loss_critic += lambda_gp * gp

        # Backprop and optimize critic
        loss_critic.backward()
        optimizer_c.step()

        # Train Generator every 5 iterations
        if i % 5 == 0:
            optimizer_g.zero_grad()

            # Generate new fake images for generator training
            z = torch.randn(real_images.size(0), z_dim).to(device)
            fake_images = generator(z)

            # Get critic score for fake images
            critic_fake = critic(fake_images)

            # Generator loss (maximize critic's score for fake images)
            loss_generator = -critic_fake.mean()

            # Backprop and optimize generator
            loss_generator.backward()
            optimizer_g.step()

        # Log progress
        if i % 100 == 0:
            print(f"Epoch [{epoch}/{epochs}], Step [{i}/{len(train_loader)}], "
                  f"Loss D: {loss_critic.item()}, Loss G: {loss_generator.item()}")

    # Save generated images
    save_image(fake_images.data[:25], f'images/fake_images_epoch_{epoch}.png', nrow=5, normalize=True)

# Display some generated images
sample_images = fake_images.data[:25].cpu()
grid = torchvision.utils.make_grid(sample_images, nrow=5, normalize=True)
plt.imshow(grid.permute(1, 2, 0))
plt.title(f"Generated Images at Epoch {epochs}")
plt.axis('off')
plt.show()
```
